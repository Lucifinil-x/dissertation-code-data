```python
import os
import pandas as pd


```

微调Public transport分类

Bus stop:1059

Underground entrance:10570794

Railway station:10570738

Taxi rank:10570758

```python

original_poi_df = pd.read_csv(r"D:\POI data\poi_cleaned_data\poi_deleted_columns.csv")
original_poi_df = original_poi_df.rename(columns={'latitude': 'lat'})
original_poi_df = original_poi_df.rename(columns={'longitude': 'lon'})
```

```python
def classify_pointx_class(num):
    str_num = str(num)
    if str_num.startswith('1059') :
        return 'Bus stop'
    elif str_num.startswith('10570794') or str_num.startswith('10570761') :
        return 'Underground entrance'
    elif str_num.startswith('10570738') :
        return 'Railway stations'
    elif str_num.startswith('10570758') :
        return 'Taxi rank'
    else:
        return 'other'

# 使用apply函数应用上述函数到每一行的'pointx_class'列，然后将结果保存到新的'poi_class'列中
original_poi_df['poi_class'] = original_poi_df['pointx_class'].apply(classify_pointx_class)
poi_change7acc = original_poi_df
poi_change7acc
```

```python
poi_change7acc_class = poi_change7acc['poi_class'].unique()
poi_change7acc_class.size
poi_change7acc_class
```

```python

dummies = pd.get_dummies(poi_change7acc['poi_class'])
poi_change7acc = pd.concat([poi_change7acc, dummies], axis=1)

for column in dummies.columns:
    poi_change7acc[column] = poi_change7acc.apply(lambda row: 1 if row['poi_class'] == column else 0, axis=1)

```

```python
poi_change7acc.columns
```

```python
poi_change7acc = poi_change7acc.drop(columns=['other'])

poi_change7acc
```

poi_change7acc删掉列Bodies of water，	Culture，	Entertainment，	Green park，	Public infrastructure，	Sports，	Tourism的值全为0的行

```python
columns_to_check = [ 'Bus stop', 'Taxi rank', 'Railway stations',
       'Underground entrance']
poi_change7acc = poi_change7acc[~(poi_change7acc[columns_to_check] == 0).all(axis=1)]
poi_change7acc
```

```python
poi_change7acc.to_csv('D:\\POI data\\poi_cleaned_data\\814_changeclass_transportation_poi.csv', index=False)
```
