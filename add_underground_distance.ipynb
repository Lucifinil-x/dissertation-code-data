```python
import pandas as pd

# 加载CSV文件
df = pd.read_csv("D:\\POI data\\poi_cleaned_data\\728_changeclass_transportation_poi.csv")

# 筛选出'Underground entrance'为1的行
filtered_df = df[df['Underground entrance'] == 1]
filtered_df

```

```python
from haversine import haversine

# 加载两个CSV文件
df1 = pd.read_csv("D:\\POI data\\729_acc\\729reclass3_add_distanceToCenter.csv")

# 定义一个函数来计算每个点到filtered_df中的所有点的最短距离
def min_distance(row):
    lat_lon = (row['lat'], row['lon'])
    distances = [haversine(lat_lon, (lat, lon)) for lat, lon in zip(filtered_df['lat'], filtered_df['lon'])]
    return min(distances)

# 应用该函数到df1的每一行
df1['distance_to_underground_entrance'] = df1.apply(min_distance, axis=1)

# 查看结果
print(df1[['lat', 'lon', 'distance_to_underground_entrance']])
```
