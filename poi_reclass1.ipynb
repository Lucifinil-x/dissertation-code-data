original_poi_df['pointx_class']里都是编号，例如3170240.,  3170245.,  6340457.,  6340454.,  3190257。我需要对df增加一列'poi_class',里面填每一行的分类属性。 下面是所有分类属性，格式为 分类属性：从左开始匹配的编号。意思是，只要这个数字匹配到编号从左开始的几位数，那么这一行的分类属性就是它；如果没有匹配的分类属性，那么就填为other. 

green park:318;
Bodies of water:358;
Culture：317;
Tourism:316,319,320; 
sports:423,424;
entertainment:421,422,425;
Public infrastructure： 6

```python

original_poi_df = pd.read_csv(r"D:\POI data\poi_cleaned_data\poi_deleted_columns.csv")
original_poi_df = original_poi_df.rename(columns={'latitude': 'lat'})
original_poi_df = original_poi_df.rename(columns={'longitude': 'lon'})
```

```python
def classify_pointx_class(num):
    str_num = str(num)
    if str_num.startswith('318'):
        return 'Green park'
    elif str_num.startswith('358'):
        return 'Bodies of water'
    elif str_num.startswith('317'):
        return 'Culture'
    elif str_num.startswith('316') or str_num.startswith('319') or str_num.startswith('320'):
        return 'Tourism'
    elif str_num.startswith('423') or str_num.startswith('424'):
        return 'Sports'
    elif str_num.startswith('421') or str_num.startswith('422') or str_num.startswith('425'):
        return 'Entertainment'
    elif str_num.startswith('6'):
        return 'Public infrastructure'
    else:
        return 'other'

# 使用apply函数应用上述函数到每一行的'pointx_class'列，然后将结果保存到新的'poi_class'列中
original_poi_df['poi_class'] = original_poi_df['pointx_class'].apply(classify_pointx_class)
poi_change7acc = original_poi_df
poi_change7acc
```

```python
poi_change7acc_class = poi_change7acc['poi_class'].unique()
poi_change7acc_class.size
poi_change7acc_class
```

```python

dummies = pd.get_dummies(poi_change7acc['poi_class'])
poi_change7acc = pd.concat([poi_change7acc, dummies], axis=1)

for column in dummies.columns:
    poi_change7acc[column] = poi_change7acc.apply(lambda row: 1 if row['poi_class'] == column else 0, axis=1)

```

```python
poi_change7acc.columns
```

```python
poi_change7acc = poi_change7acc.drop(columns=['other'])

poi_change7acc
```

poi_change7acc删掉列Bodies of water，	Culture，	Entertainment，	Green park，	Public infrastructure，	Sports，	Tourism的值全为0的行

```python
columns_to_check = ['Bodies of water', 'Culture', 'Entertainment', 'Green park', 'Public infrastructure', 'Sports', 'Tourism']
poi_change7acc = poi_change7acc[~(poi_change7acc[columns_to_check] == 0).all(axis=1)]
poi_change7acc
```
