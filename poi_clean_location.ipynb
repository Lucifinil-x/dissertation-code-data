把文件"D:\POI data\london\Download_london_first_poi_csv_2292170\poi_5097912\poi-extract-2023_06.csv"用python进行清洗，feature_easting列是503981，feature_northing是171769。将这两列转化为对应的经纬度作为新列

```python
from pyproj import Proj, transform
import pandas as pd
import pyproj
# 定义转换函数
def convert_to_lat_lng(row):
    easting = row['feature_easting']
    northing = row['feature_northing']
    
    # 定义转换器
    transformer = pyproj.Transformer.from_crs('epsg:27700', 'epsg:4326', always_xy=True)
    
    # 执行转换
    longitude, latitude = transformer.transform(easting, northing)
    return latitude, longitude

# 读取CSV文件
file_path = r"D:\POI data\london\Download_london_accomodation_csv_2292171\poi_5097913\poi-extract-2023_06.csv"
df = pd.read_csv(file_path)

# 创建新列"latitude"和"longitude"，并应用转换函数
df[['latitude', 'longitude']] = df.apply(convert_to_lat_lng, axis=1, result_type='expand')

# 打印结果
df

# 保存到新的 CSV 文件
df.to_csv(r"D:\s-casa毕业论文-工程\R5R_TEST1_london\poi_accomodation", index=False)


```
