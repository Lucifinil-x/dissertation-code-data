```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

def testResult(ann_model):
    # 预测训练集和测试集
    y_train_pred = ann_model.predict(X_train).flatten()
    y_test_pred = ann_model.predict(X_test).flatten()


    # 计算均方根误差（RMSE）作为性能指标
    train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
    test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))

    #print(f"Train RMSE: {train_rmse}")
    print(f"Test RMSE: {test_rmse}")

    # 计算训练集和测试集的R²值
    train_r2 = r2_score(y_train, y_train_pred)
    test_r2 = r2_score(y_test, y_test_pred)

    #print(f"Train R²: {train_r2}")
    print(f"Test R²: {test_r2}")

    # 计算训练集和测试集的MAPE值
    train_mape = np.mean(np.abs((y_train - y_train_pred) / y_train)) * 100
    test_mape = np.mean(np.abs((y_test - y_test_pred) / y_test)) * 100

    #print(f"Train MAPE: {train_mape:.2f}%")
    print(f"Test MAPE: {test_mape:.2f}%")
    
```

```python
# 获取特征数量
n_features = X_train.shape[1]

# 初始化ANN模型
model = Sequential()

# 添加输入层和第一个隐藏层，添加dropout防止过拟合
model.add(Dense(units=128, activation='relu', input_shape=(n_features,)))
model.add(tf.keras.layers.Dropout(0.2))

# 添加第二个隐藏层，添加dropout防止过拟合
model.add(Dense(units=64, activation='relu'))
model.add(tf.keras.layers.Dropout(0.2))

# 添加第三个隐藏层，添加dropout防止过拟合
model.add(Dense(units=32, activation='relu'))
model.add(tf.keras.layers.Dropout(0.2))

# 添加输出层
model.add(Dense(units=1))

# 编译模型
model.compile(optimizer='rmsprop', loss='mean_squared_logarithmic_error')

# 训练模型
model.fit(X_train, y_train, epochs=200, batch_size=32)
testResult(model)
```
