```python
from pysal.lib import weights
from pysal.model import spreg

# Combine 'lat' and 'lon' into a 2D numpy array
coords = X[['lon', 'lat']].values
# Create a KNN spatial weight
w = weights.KNN.from_array(coords, k=5)
w.transform = 'r'  # row-standardization

# Convert X and y to numpy arrays
X_np = X.values
y_np = y.values.reshape((-1, 1))

# Run the spatial error model
model = spreg.GM_Error(y_np, X_np, w=w, name_x=X.columns.tolist(), name_y='Price_per_square_meter')

# Print the summary
print(model.summary)
```
