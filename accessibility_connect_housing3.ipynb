```python
import pandas as pd
import os

# 读取 'housing_hasId.csv' 文件
df_main = pd.read_csv(r"D:\POI data\poi_cleaned_data\728_original_9acc_housing_Id.csv")

# 获取所有以 '15m.csv' 结尾的文件
csv_files = [f for f in os.listdir(r"D:\POI data\728_7acc\reclass2") if f.endswith('15m.csv')]

# 对每个 '15m.csv' 文件进行操作
for csv_file in csv_files:
    # 读取文件
    df = pd.read_csv(os.path.join(r"D:\POI data\728_7acc\reclass2", csv_file))
    
    # 新增一列，列名是文件名，值为 'accessibility' 列的值
    new_column_name = os.path.splitext(csv_file)[0]  # 去除文件扩展名
    df[new_column_name] = df['accessibility']
    
    # 根据 'id' 列连接 DataFrame，只添加新的列
    df_main = pd.merge(df_main, df[['id', new_column_name]], on='id')

# 打印结果 DataFrame
df_main
```

```python
df_main['tenure'] = df_main['tenure'].fillna('Unknown')

df_main.to_csv(r"D:\POI data\728_7acc\housing_hasId_reclass2_acc.csv", index=False)
```
